> **Write a program that takes as input a BST and an interval and returns the BST keys that lie in the interval. For example, for the tree in Figure 14.1 on Page 198, and interval [16,31], you should return 17, 19, 23, 29, 31.**

_Hint: How many edges are traversed when the successor function is repeatedly called m times?_



```python
from typing import List, Optional
from collections import namedtuple

class BstNode:
    def __init__(self, data=None, left=None, right=None):
        self.data, self.left, self.right = data, left, right

    def __eq__(self, other):
        return equal_binary_trees(self, other)

    def __repr__(self):
        return str(binary_tree_to_string(self))

    def __str__(self):
        return self.__repr__()

Interval = namedtuple('Interval', ('left', 'right'))

def range_lookup_in_bst(tree: BstNode, interval: Interval) -> List[int]:
    def tree_look_up(tree, interval, result):
        if not tree:
            return None
        
        if interval.left <= tree.data <= interval.right:
            tree_look_up(tree.left, interval, result)
            result.append(tree.data)
            tree_look_up(tree.right, interval, result)
        elif interval.left > tree.data:
            tree_look_up(tree.right, interval, result)
        else: #tree.data < interval.right:
            tree_look_up(tree.left, interval, result)

    return tree_look_up(tree, interval, [])
```
