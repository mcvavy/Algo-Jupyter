```python
from typing import List

"""

"""
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        maxArea = 0
        stack = [] # pair of (index, height)
        
        for i, h in enumerate(heights):
            start = i
            while stack and stack[-1][1] > h:
                index, height = stack.pop()
                maxArea = max(maxArea, height * (i - index))
                start = index
            stack.append((start, h))
            
        for i, h in stack:
            maxArea = max(maxArea, h * (len(heights) - i))
        return maxArea
    
import unittest
class LargestRectangleAreaTest(unittest.TestCase):
    def test_largestRectangleArea_test1(self):
        result = Solution().largestRectangleArea([2,1,5,6,2,3])
        self.assertEqual(result, 10)
        
    def test_largestRectangleArea_test2(self):
        result = Solution().largestRectangleArea([2,4])
        self.assertEqual(result, 4)
        
unittest.main(argv=[''], verbosity=2, exit=False)
```
