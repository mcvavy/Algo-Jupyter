> https://leetcode.com/problems/container-with-most-water/

  

- [TheCodingWorld Explanation](https://www.youtube.com/watch?v=wmZEcCoHgZU&ab_channel=thecodingworld)  

_Time complexity: O(N) | Space complexity: O(1)_

```python
from typing import List

class Solution:
    def maxArea(self, height: List[int]) -> int:
        if not height:
            return 0
        
        maximum_area = float("-inf")
        start = 0; end = len(height) - 1
        
        while start < end:
            l = min(height[start], height[end]); b = end - start
            
            tempMaxArea = l * b
            maximum_area = max(tempMaxArea, maximum_area)
            
            if height[start] < height[end]:
                start += 1
            else:
                end -= 1
        return maximum_area
s = Solution()
s.maxArea([1,8,6,2,5,4,8,3,7])
```
