> https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/

> Optimized Solution  



```python
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        start = 0; end = len(nums) - 1
        
        while start < end:
            mid = (start + end)//2
            
            if nums[mid] >= nums[end]:
                start = mid + 1
            else:
                end = mid
        return nums[start]
s = Solution()
s.findMin([4,5,6,7,0,1,2])
```
