> Brute Force Approach

_Time complexity: $O(N^2)$  | Space complexity: O(1)_

```python
from typing import List

class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        result = nums[0]
        
        for i in range(len(nums)):
            product = 1
            for j in range(i, len(nums)):
                product = product * nums[j]
                result = max(product, result)
        return result
s = Solution()
s.maxProduct([-2,1,-3,4,-1,2,1,-5,4])
```

> Optimized Solution

```python
from typing import List

class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        maximum = current_max = current_min = nums[0]
        
        for i in nums[1:]:
            temp = current_max
            current_max = max(i, current_max * i, current_min * i)
            current_min = min(i, temp*i, current_min * i)
            maximum = max(current_max, maximum)
        return maximum
s = Solution()
s.maxProduct([-2,1,-3,4,-1,2,1,-5,4])
```
