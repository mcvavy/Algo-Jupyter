> **4 pointer approach**

_Time complexity: $O(N^2)$ | Space complexity: O(N)_



- [Xavier Elon Explanation](https://www.youtube.com/watch?v=ViWmh5Eq8xk&ab_channel=XavierElon)

```python
from typing import List

class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:     
        res = []
        length = len(nums)
        
        if not nums or length < 4:
            return res
        
        nums.sort()
        for i in range(length - 3):
            if i > 0 and nums[i] == nums[i-1]:
                continue
            for j in range(i + 1, length - 2):
                if j > i + 1 and nums[j] == nums[j-1]:
                    continue
                l = j + 1; r = length - 1
                while l < r:
                    current_sum = nums[i] + nums[j] + nums[l] + nums[r]
                    if current_sum == target:
                        res.append([nums[i] , nums[j] , nums[l] , nums[r]])
                        while l<r and nums[l] == nums[l + 1]:
                            l += 1
                        while l<r and nums[r] == nums[r-1]:
                            r -= 1
                        l += 1
                        r -= 1
                    elif current_sum > target:
                        r -= 1
                    else:
                        l += 1
        return res
s = Solution()
s.fourSum([1,0,-1,0,-2,2], 0)
```

> **HashMap Approach**

_Time complexity: $O(N^2)$ | Space complexity: O(N)_



- [Xavier Elon Explanation](https://www.youtube.com/watch?v=ViWmh5Eq8xk&ab_channel=XavierElon)

```python
from typing import List

class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:     
        res = []
        hm = {}
        
        n = len(nums)
        for outer in range(n):
            for right in range(outer + 1, n):
                currentSum = nums[outer] + nums[right]
                currentDiff = target - currentSum
                if currentDiff in hm:
                    for pair in hm.get(currentDiff):
                        res.append(pair + [nums[outer], nums[right]])
                        
            for left in range(0, outer):
                tempSum = nums[left] + nums[outer]
                if tempSum in hm:
                    # Update the current value with key tempSum
                    hm[tempSum].append([nums[left], nums[outer]])
                else:
                    hm[tempSum] = [[nums[outer], nums[left]]]
        return res
                
                
s = Solution()
s.fourSum([7, 6, 4, -1, 1, 2], 16)
```
