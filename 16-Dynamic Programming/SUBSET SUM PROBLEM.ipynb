```python
import numpy as np

# Time Complexity: O(sum*n), where sum is the ‘target sum’ and ‘n’ is the size of array.
# Auxiliary Space: O(sum*n), as the size of 2-D array is sum*n. 

def subSetSum(A, target):
    dp = [[True if x == 0 else False for x in range(target + 1)] for y in range(len(A) + 1)]
    print(np.matrix(dp), end= " \n\n")
    for row in range(1, len(A) + 1):
        for col in range(1,target + 1):
            if col < A[row - 1]:
                dp[row][col] = dp[row - 1][col]
            else:
                dp[row][col] = dp[row - 1][col] or dp[row - 1][col - A[row - 1]]
    print(np.matrix(dp))
    return dp[len(A)][target]
subSetSum([2,3,7,8,10], 11)
    
```

> Recursion approach

```python
def subSetSum(A, target):
    def helper(arr, n, target):
        if target == 0:
            return True
        if n == 0:
            return False
        if arr[n-1] > target:
            return helper(arr, n-1, target)
        
        return helper(arr, n-1, target) or helper(arr, n-1, target - arr[n-1])
    return helper(A, len(A) - 1, target)
subSetSum([2,3,7,8,10], 11)
```
