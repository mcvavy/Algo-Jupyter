>**Coin Change Problem**



<br>
<br>



<br>
<br>

_Time Complexity: O(M*N) | Space complexity: O(M*N)_


```python
from typing import List

class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        """
                V\A  0   1   2   3   4   5   6
                   +---+---+---+---+---+---+---+
                0  | 0 | I | I | I | I | I | I |
                   +---+---+---+---+---+---+---+
                1  | 0 |   |   |   |   |   |   |
                   +---+---+---+---+---+---+---+
                2  | 0 |   |   |   |   |   |   |
                   +---+---+---+---+---+---+---+
                3  | 0 |   |   |   |   |   |   |
                   +---+---+---+---+---+---+---+
        """
        dp = [[0 for _ in range(amount + 1)] for _ in range(len(coins) + 11)]
        
        for i in range(1, len(dp[0])):
            dp[0][i] = float("inf")
        
        for i in range(1, len(coins) + 1):
            for j in range(1, amount + 1):
                if coins[i - 1] > j:
                    # Exclusion
                    dp[i][j] = dp[i-1][j]
                else:
                    dp[i][j] = min(dp[i-1][j], 1 + dp[i][j - coins[i-1]])
        return -1 if dp[len(coins)][amount] == float("inf") else dp[len(coins)][amount]
        


import unittest
class CoinChangeTest(unittest.TestCase):
    def test_coinChange_test1(self):
        result = Solution().coinChange([1,2,5], 11)
        self.assertEqual(result, 3)

    def test_coinChange_test2(self):
        result = Solution().coinChange([2], 3)
        self.assertEqual(result, -1)
        
    def test_coinChange_test3(self):
        result = Solution().coinChange([1], 1)
        self.assertEqual(result, 1)
        
unittest.main(argv=[''], verbosity=2, exit=False)
        
```
