> **Recursion + memoization**  

- As the input get becomes larger, the recusion depth will max out. This is not an optimal solution

```python
from typing import List
import numpy

class Solution:
    def minOperations(self, nums: List[int], x: int) -> int:
        def reduce(nums, x, l, r, count, memo):
            if x == 0:
                return count
            elif x < 0 or l > r:
                return float("inf")
            else:
                key = str(l) + "*" + str(r) + "*" + str(x)
                if key in memo:
                    return memo[key]
                leftReduce = reduce(nums, x - nums[l], l + 1, r, count + 1, memo)
                rightReduce = reduce(nums, x - nums[r], l, r - 1, count + 1, memo)
                minimum = min(leftReduce, rightReduce)
                memo[key] = minimum
                return minimum
        L = 0
        R = len(nums) - 1
        memo = {}
        minOperation = reduce(nums, x, L, R, 0, memo)
        return minOperation if minOperation != float("inf") else -1


import unittest
class MinimumOperationTest(unittest.TestCase):
    def test_minOperations_test1(self):
        result = Solution().minOperations([1,1,4,2,3], 5)
        self.assertEqual(result, 2)

    def test_minOperations_test2(self):
        result = Solution().minOperations([5,6,7,8,9], 4)
        self.assertEqual(result, -1)
        
    def test_minOperations_test3(self):
        result = Solution().minOperations([3,2,20,1,1,3], 10)
        self.assertEqual(result, 5)
        
unittest.main(argv=[''], verbosity=2, exit=False)
```
