```python
from typing import List

class Solution:
    def maxSubarraySumCircular(self, A: List[int]) -> int:
        total_sum = 0
        max_sum = float("-inf"); current_max = 0
        min_sum = float("inf"); current_min = 0
        
        for x in A:
            total_sum += x
            
            current_max = max(current_max + x, x)
            max_sum = max(max_sum, current_max)
            
            current_min = min(current_min + x, x)
            min_sum = min(min_sum, current_min)
        
        if max_sum > 0:
            return max(total_sum - min_sum, max_sum)
        return max_sum
            
            
            
                    
import unittest
class MaxCircularSunArraySumTest(unittest.TestCase):
    def test_max_sum_circular_test1(self):
        result = Solution().maxSubarraySumCircular([1,-2,3,-2])
        self.assertEqual(result, 3)
        
    def test_max_sum_circular_test2(self):
        result = Solution().maxSubarraySumCircular([5,-3,5])
        self.assertEqual(result, 10)
                        
    def test_max_sum_circular_test3(self):
        result = Solution().maxSubarraySumCircular([3,-1,2,-1])
        self.assertEqual(result, 4)
        
    def test_max_sum_circular_test4(self):
        result = Solution().maxSubarraySumCircular([3,-2,2,-3])
        self.assertEqual(result, 3)
        
    def test_max_sum_circular_test5(self):
        result = Solution().maxSubarraySumCircular([-2,-3,-1])
        self.assertEqual(result, -1)
        
unittest.main(argv=[''], verbosity=2, exit=False)
```
