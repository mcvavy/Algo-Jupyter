```python
def is_match(text, pattern):
    p_len = len(pattern)
    t_len = len(text)
    
    dp = [[False for _ in range(p_len + 1)] for _ in range(t_len + 1)]
    dp[0][0] = True
    
    for i in range(1, len(dp[0])):
        if pattern[i-1] == "*":
            dp[0][i] = dp[0][i-2]
    
    for i in range(1, t_len + 1):
        for j in range(1, p_len + 1):
            if text[i-1] == pattern[j - 1] or pattern[j-1] == ".":
                dp[i][j] = dp[i-1][j-1]
            elif pattern[j-1] == "*":
                dp[i][j] = dp[i][j-2]
                if pattern[j-2] == "." or pattern[j-2] == text[i-1]:
                    dp[i][j] = dp[i][j] or dp[i-1][j] 
            else:
                dp[i][j] = False
    return dp[t_len][p_len]
                
                



import unittest
class IsMatchTest(unittest.TestCase):
    def test_is_match_test1(self):
        result = is_match( "aa", "a")
        self.assertEqual(result, False)

    def test_is_match_test2(self):
        result = is_match( "aa", "aa")
        self.assertEqual(result, True)

    def test_is_match_test3(self):
        result = is_match("abc", "a.c")
        self.assertEqual(result, True)
        
    def test_is_match_test4(self):
        result = is_match("abbb", "ab*")
        self.assertEqual(result, True)
        
    def test_is_match_test5(self):
        result = is_match("acd", "ab*c.")
        self.assertEqual(result, True)
        
unittest.main(argv=[''], verbosity=2, exit=False)
    
```
