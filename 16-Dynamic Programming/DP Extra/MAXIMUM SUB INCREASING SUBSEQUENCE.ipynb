```python
import numpy
from typing import List

class Solution:
    def maxSumIncreasingSubsequence(self, sequence: List[int]) -> int:
        msis = sequence.copy()
        
        for i in range(len(sequence)):
            for j in range(i):
                if sequence[i] > sequence[j]:
                    msis[i] = max(msis[i], sequence[i] + msis[j])
                    
        return max(msis)
    

import unittest
class MaxEnvelopeTest(unittest.TestCase):
    def test_lis_test1(self):
        result = Solution().maxSumIncreasingSubsequence([1, 202, 2, 3, 100, 4, 5])
        self.assertEqual(result, 3)
        
unittest.main(argv=[''], verbosity=2, exit=False)
```
