>**Pattern Matching**

Given a pattern P and a string S, find is P occurs as a subsequence in S.

- [Pattern Matching](https://www.youtube.com/watch?v=eG5NP5KBJVA&list=PLEJXowNB4kPxBwaXtRO1qFLpCzF75DYrS&index=23&ab_channel=TECHDOSE)

Solution:

    1. Find LCS of S and P and if P equals LCS, then P occurs in S
<br>


<br>


    

```python
import numpy

class Solution:
    """
    DP tabulation solution (bottom-up approach)
    Time complexity: O(M*N)
    """
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        m = len(text2); n = len(text1);
        dp = [["" for _ in range(m + 1)] for _ in range(n + 1)]

        for i in range(1,n + 1):
            for j in range(1, m + 1):
                if text1[i - 1] == text2[j-1]:
                    dp[i][j] = dp[i-1][j-1] + text2[j-1]
                else:
                    dp[i][j] = dp[i-1][j] if len(dp[i-1][j]) > len(dp[i][j-1]) else dp[i][j-1]
        print(numpy.matrix(dp), end="\n\n")
        return dp[n][m]

    def shortestCommonSupersequence(self, str1: str, str2: str) -> str:
        result = ""
        lcs = self.longestCommonSubsequence(str1, str2)
        
        p1 = 0; p2 = 0
        for c in lcs:
            while str1[p1] != c:
                result += str1[p1]; p1 += 1
            while str2[p2] != c:
                result += str2[p2]; p2 += 1
            result += c
            p1 += 1; p2 += 1
            
        result += str1[p1:] + str2[p2:]
        return result
    

import unittest
class LCSTest(unittest.TestCase):
    def test_lcs_test1(self):
        result = Solution().shortestCommonSupersequence("abcde","ace")
        self.assertEqual(result, "abcde")
        
    def test_lcs_test2(self):
        result = Solution().shortestCommonSupersequence("abc","abc")
        self.assertEqual(result, "abc")
        
    def test_lcs_test3(self):
        result = Solution().shortestCommonSupersequence("abac", "cab")
        self.assertEqual(result, "cabac")
        
unittest.main(argv=[''], verbosity=2, exit=False)
```
