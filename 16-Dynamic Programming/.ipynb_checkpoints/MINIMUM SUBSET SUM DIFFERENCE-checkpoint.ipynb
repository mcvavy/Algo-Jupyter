1. Consider that column 0 above should all be 1s, because it will take an item (emppty subset) to form 0 
2.
    $f(x)= \begin{cases}
    F(i-1, j),& \text{if } A(i-1, j) > j\\
    F(i,j) = F(i-1, j) + F(i-1, j-A(i-1)),& \text{else}
\end{cases}
$



```python
from typing import List
import numpy

def count_subset_with_sum(arr, targetSum) -> int:
    n = len(arr)
    
    dp = [[0 for _ in range(targetSum + 1)] for _ in range(n + 1)]
    
    """
    - To get a sum of 0, one item can be used which is the empty set
    - Same for first row to be zero as zero cannot form sums starting from index 1
    """
    for i in range(n + 1):
        dp[i][0] = 1
    print(numpy.matrix(dp), end="\n\n")
    
    for i in range(1, n + 1):
        for j in range(1, targetSum + 1):
            if arr[i-1] > j:
                dp[i][j] = dp[i - 1][j]
            else:
                dp[i][j] = dp[i-1][j] + dp[i-1][j - arr[i-1]]
    return dp[n][targetSum]



import unittest
class CountSubSetSumTest(unittest.TestCase):
    def test_cocount_subset_with_sum_test1(self):
        A = [1,2,3]
        result = count_subset_with_sum(A, 3)
        self.assertEqual(result, 2)
        
    def test_cocount_subset_with_sum_test2(self):
        A = [3,1,2,3]
        result = count_subset_with_sum(A, 6)
        self.assertEqual(result, 3)

unittest.main(argv=[''], verbosity=2, exit=False)
    
```
