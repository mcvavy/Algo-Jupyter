>**Coin Change 2 Problem**  



```python
from typing import List
import numpy

class Solution:
    def change(self, amount: int, coins: List[int]) -> int:
        """
                V\A  0   1   2   3   4   5   6
                   +---+---+---+---+---+---+---+
                0  | 1 | 0 | 0 | 0 | 0 | 0 | 0 |
                   +---+---+---+---+---+---+---+
                1  | 1 |   |   |   |   |   |   |
                   +---+---+---+---+---+---+---+
                2  | 1 |   |   |   |   |   |   |
                   +---+---+---+---+---+---+---+
                3  | 1 |   |   |   |   |   |   |
                   +---+---+---+---+---+---+---+
        """
        n = len(coins)
        dp = [[0 for _ in range(amount + 1)] for _ in range(n + 1)]

        for row in range(len(dp)):
            dp[row][0] = 1

        for i in range(1, n + 1):
            for j in range(1, amount + 1):
                if coins[i - 1] > j:
                    dp[i][j] = dp[i-1][j]
                else:
                    dp[i][j] = dp[i-1][j] + dp[i][j - coins[i-1]]
        return dp[n][amount]
        



import unittest
class CoinChangeTwoTest(unittest.TestCase):
    def test_coinChange2_test1(self):
        result = Solution().change(5, [1, 2, 5])
        self.assertEqual(result, 4)

    def test_coinChange2_test2(self):
        result = Solution().change(3, [2])
        self.assertEqual(result, 0)
        
    def test_coinChange2_test3(self):
        result = Solution().change(10, [10])
        self.assertEqual(result, 1)
        
unittest.main(argv=[''], verbosity=2, exit=False)
```
