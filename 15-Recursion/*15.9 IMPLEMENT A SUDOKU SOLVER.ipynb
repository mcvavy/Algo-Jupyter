```python
from typing import List
import numpy as np

def solve_sudoku(partial_assignment: List[List[int]]) -> bool:
    def find_empty(grid):
        for row in range(len(grid)):
            for column in range(len(grid[0])):
                if grid[row][column] == 0:
                    return (row, column)
        return None

    def isCellPossible(grid, row, column, val):
        # Check row
        for col in range(9):
            if grid[row][col] == val:
                return False
        # Check Column
        for rw in range(9):
            if grid[rw][column] == val:
                return False
        # Check cell
        row0 = (row//3)*3
        col0 = (column//3)*3

        for r in range(3):
            for c in range(3):
                if grid[r+row0][c+col0] == val:
                    return False
        return True
    
    def bt(grid):
#         print(np.matrix(grid))
        find = find_empty(grid)

        if not find:
            return True
        else:
            row, column = find

        for i in range(1, len(grid) + 1):
            if isCellPossible(grid, row, column, i):
                grid[row][column] = i

                if bt(grid):
                    return True
                grid[row][column] = 0
        return False

    return bt(partial_assignment)
A = [[0, 3, 2, 0, 0, 0, 8, 0, 4], 
     [8, 0, 0, 2, 0, 0, 0, 7, 0], 
     [0, 1, 7, 0, 0, 5, 9, 0, 6], 
     [5, 8, 0, 0, 2, 0, 0, 3, 0], 
     [0, 0, 6, 0, 4, 0, 7, 0, 0], 
     [0, 0, 4, 9, 1, 3, 0, 6, 0], 
     [0, 0, 0, 7, 3, 0, 2, 0, 0], 
     [0, 5, 9, 0, 0, 0, 0, 0, 1], 
     [1, 0, 0, 8, 0, 9, 0, 0, 0]]

solve_sudoku(A)
A
```

```python
from typing import List

def solveSudoku(board: List[List[str]]) -> None:
    def findSpace(grid):
        for row in range(len(grid)):
            for col in range(len(grid[0])):
                if grid[row][col] == ".":
                    return (row, col)
        return None

    def isValidCell(row, col, grid, val):
        #Check row
        for c in range(9):
            if grid[row][c] == val:
                return False

        #Chec column
        for r in range(9):
            if grid[r][col] == val:
                return False

        #Check cell
        r0 = (row//3) * 3
        c0 = (col//3) * 3

        for i in range(3):
            for j in range(3):
                if grid[r0+i][c0+j] == val:
                    return False
        return True

    def bt(grid):
        space = findSpace(grid)
        if not space:
            return True
        else:
            row, column = space
        for val in range(1, 10):
            if isValidCell(row,column,grid, str(val)):
                grid[row][column] = str(val)
                if bt(grid):
                    return True
                grid[row][column] = "."
        return False
    bt(board)
    
A = [["5","3",".",".","7",".",".",".","."],
     ["6",".",".","1","9","5",".",".","."],
     [".","9","8",".",".",".",".","6","."],
     ["8",".",".",".","6",".",".",".","3"],
     ["4",".",".","8",".","3",".",".","1"],
     ["7",".",".",".","2",".",".",".","6"],
     [".","6",".",".",".",".","2","8","."],
     [".",".",".","4","1","9",".",".","5"],
     [".",".",".",".","8",".",".","7","9"]]

solveSudoku(A)
A
```
