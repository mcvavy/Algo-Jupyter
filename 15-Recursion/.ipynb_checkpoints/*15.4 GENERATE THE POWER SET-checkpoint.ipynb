

> **Write a function that takes as input a set and retums its power set.**  

_Hint: There are $2^n$ subsets for a given set S of size n. There are $2^k$ k-bit words._

- [Solution Explanation](https://www.youtube.com/watch?v=1BbeOUkUOak&ab_channel=KnowledgeCenter)

> **Recursive Solution**

- [Youtube Explanation](https://www.youtube.com/watch?v=LiuEfq37hqc&t=69s&ab_channel=SuboptimalEngineer)

```python
from typing import List

def generate_power_set(nums: List[int]) -> List[List[int]]:
    def generate(nums, result, curr, index):
        result.append(list(curr))
        for i in range(index,len(nums)):
            curr.append(nums[i])
            generate(nums, result, curr, i + 1)
            curr.pop()
    
    result = []
    generate(nums, result, [], 0)
    return result

generate_power_set([1,2,3])
```

> **Recursive Backtracking**

```python
from typing import List

def generate_power_set(nums: List[int]) -> List[List[int]]:
    def bt(sublist, pos):
        res.append(list(sublist))
        for i in range(pos, len(nums)):
            sublist.append(nums[i])
            bt(sublist, i + 1)
            sublist.pop()
    res = []
    bt([], 0)
    return res

generate_power_set([1,2,3])
```

> **Iterative Solution**  

Time complexity: O(N x $2^N$)  | Space complexity: O(N x $2^N$)

```python
from typing import List


def generate_power_set(input_set: List[int]) -> List[List[int]]:
    if not input_set: return [[]]
    result = [[]]
    
    for x in input_set:
        n = len(result)
        for i in range(n):
            r = result[i] + [x]
            result.append(r)
    
    return result

generate_power_set([1,2,3])
```
