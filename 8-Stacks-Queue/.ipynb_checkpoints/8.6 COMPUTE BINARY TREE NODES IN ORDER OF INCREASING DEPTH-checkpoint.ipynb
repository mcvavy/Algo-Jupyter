> **Given a binary tree, return an array consisting of the keys at the same level. Keys should appear in the order of the corresponding nodes' depths, breaking ties from left to right. For example, you should return [(314), (6,6), (271,561,2,271), (28,0,3,1,28), (17,401,257), (641)] for the binary tree in Figure 9.1 on Page112.**   


_Hint: First think about solving this problem with a pair of queues._    





```python
def binary_tree_depth_order(tree: BinaryTreeNode) -> List[List[int]]:
    if not tree:
        return []
    
    res = []
    queue = [tree]
    
    while queue:
        res.append([node.data for node in queue])
        
        queue = [child for curr in queue for child in [curr.left, curr.right] if child]
    
    return res
```
