> **Write a program that takes a perfect binary tree, and sets each node's level-next field to the node on its right, if one exists.**  

_Hint: Think of an appropriate traversal order._  



- [Solution Explanation](https://www.youtube.com/watch?v=ZdDs2ps7rEs&ab_channel=CodingSimplified)

```python


class BinaryTreeNode:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
        self.next = None  # Populates this field.

from collections import deque
def construct_right_sibling(tree: BinaryTreeNode) -> None:
    if not tree: return 
    q = deque()
    q.append(tree)
    
    while q:
        size = len(q)
        prev = None
        while size:
            node = q.popleft()
            if prev:
                prev.next = node
            prev = node
            
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
            size -= 1
    return
```
