> **Design an algorithm that takes a 2D sorted array and a number and checks whether that number appears in the array. For example, if the input is the 2D sorted array in Figure 11.3, and the number is 7, your algorithm should return false; if the number is 8, your algorithm should return true.**   

_Hint: Can you eliminate a row or a column per comparison?_   



- [Video Walkthrough Explanation](https://www.youtube.com/watch?v=dsPdwhRR_84&ab_channel=VivekanandKhyade-AlgorithmEveryDay)  

 



```python
from typing import List

def matrix_search(A: List[List[int]], x: int) -> bool:
    if not A: return False
    
    i = 0; j = len(A[0]) - 1
    while i < len(A) and j >= 0:
        if A[i][j] == x:
            return True
        elif A[i][j] > x:
            j -= 1
        else:
            i += 1
    return False
    
```
