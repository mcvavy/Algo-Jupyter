> **Design an O(logn) algorithm for finding the position of the smallest element in a cyclically sorted array. Assume all elements are distinct. For example, for the array in Figure 11.2, your algorithm should return 4.**   

_Hint: Reduce this problem to ordinary binary search._  




```python
from typing import List

# Time complexity: O(logN)
def search_smallest(A: List[int]) -> int:
    start = 0; end = len(A) - 1
    
    while start < end:
        mid = (start + end)//2
        if A[mid] > A[end]:
            start = mid + 1
        else:
            end = mid
    return start

search_smallest([378, 478, 550, 631, 103, 203, 220, 234, 279, 368])
```

> **Design an $O(\log n)$ algorithm for finding the position of an element k in a cyclically sorted array of distinct elements.**

```python
# Time complexity: O(logN)
def find_k_in_sorted(A: List[int], k) -> int:
    start = 0; end = len(A) - 1
    
    while start <= end:
        mid = (start + end)//2
        if A[mid] == k:
            return mid
        elif A[start] <= A[mid]:
            if k >= A[start] and k < A[mid]:
                end = mid - 1
            else:
                start = mid + 1
        elif A[mid] <= A[end]:
            if k > A[mid] and k <= A[end]:
                start = mid + 1
            else:
                end = mid - 1

    return -1

find_k_in_sorted([378, 478, 550, 631, 103, 203, 220, 234, 279, 368], 220)
```
