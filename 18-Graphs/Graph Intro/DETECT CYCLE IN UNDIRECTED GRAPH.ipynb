```python
from enum import Enum, auto

class State(Enum):
    UNVISITED = auto()
    INSTACK = auto()
    VISITED = auto()

class Graph:
    def __init__(self, v):
        self.n_v = v
        self.adj = [[] for _ in range(v)]
    def add_edge(self,u,v):
        self.adj[u].append(v)
        self.adj[v].append(u)
    def BFS(self, s):
        result = []
        Q = deque(); visited = [False] * self.n_v
        
        Q.append(s)
        visited[s] = True
        while Q:
            current = Q.popleft()
            result.append(current)
            for neighbours in self.adj[current]:
                if not visited[neighbours]:
                    Q.append(neighbours)
                    visited[neighbours] = True
        return result
    
    def DFS(self):
        def dfs_helper(current, visited):
            visited[current] = True
            result.append(current)
            for neighbour in self.adj[current]:
                if not visited[neighbour]:
                    dfs_helper(neighbour, visited)
        result = []
        visited = [False] * self.n_v
        
        for i in range(self.n_v):
            if not visited[i]:
                dfs_helper(i, visited)
        return result
    
    def DFS_iter(self):
        def dfs_helper_it(vertice, v):
            S = deque()
            S.append(vertice)

            while S:
                current = S.pop()
                result.append(current)

                for adj in self.adj[current]:
                    if not v[adj]:
                        S.append(adj)
                        v[adj] = True
        
        result = []
        visited = [False] * self.n_v
        for i in range(self.n_v):
            if not visited[i]:
                dfs_helper_it(i, visited)
                
        return result
    
    def has_cycle(self):
        def helper(current, v):
            v[current] = State.INSTACK
            for u in self.adj[current]:
                if v[u] == State.INSTACK:
                    return True
                if v[u] == State.UNVISITED and helper(u, visited):
                    return True
            v[current] = State.VISITED
            return False
            
        visited = [State.UNVISITED] * self.n_v
        for i in range(self.n_v):
            if visited[i] == State.UNVISITED and helper(i, visited):
                return True
        return False
    
    def has_cycle_undirected(self):
        def undirected_helper(current, v, parent):
            v[current] = True
            for u in self.adj[current]:
                if v[u] and u != parent: return True
                if not v[u] and undirected_helper(u,v, current): return True
            return False
        
        visited = [False] * self.n_v
        for i in range(self.n_v):
            if not visited[i] and undirected_helper(i, visited, -1):
                return True
        return False
        
    
G = Graph(5)
G.add_edge(0,1)
G.add_edge(0,3)
# G.add_edge(0,4)
G.add_edge(1,2)
G.add_edge(3,4)
# G.add_edge(2,4)

print("Bread FIrst Search => ", G.BFS(0))
print("Depth First Search => ", G.DFS())
print("Depth First Search Iterative", G.DFS_iter())
print(G.has_cycle())
print(G.has_cycle_undirected())
```
