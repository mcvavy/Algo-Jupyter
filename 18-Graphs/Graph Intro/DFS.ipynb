```python
from collections import deque

class Graph:
    def __init__(self, v):
        self.n_v = v
        self.adj = [[] for _ in range(v)]
    def add_edge(self,u,v):
        self.adj[u].append(v)
        # self.adj[v].append(u) -> Undirected graph
    def BFS(self, s):
        result = []
        Q = deque(); visited = [False] * self.n_v
        
        Q.append(s)
        visited[s] = True
        while Q:
            current = Q.popleft()
            result.append(current)
            for neighbours in self.adj[current]:
                if not visited[neighbours]:
                    Q.append(neighbours)
                    visited[neighbours] = True
        return result
    
    def DFS(self):
        def dfs_helper(current, visited):
            visited[current] = True
            result.append(current)
            for neighbour in self.adj[current]:
                if not visited[neighbour]:
                    dfs_helper(neighbour, visited)
        result = []
        visited = [False] * self.n_v
        
        for i in range(self.n_v):
            if not visited[i]:
                dfs_helper(i, visited)
        return result
    
    def DFS_iter(self):
        def dfs_helper_it(vertice, v):
            S = deque()
            S.append(vertice)

            while S:
                current = S.pop()
                result.append(current)

                for adj in self.adj[current]:
                    if not v[adj]:
                        S.append(adj)
                        v[adj] = True
        
        result = []
        visited = [False] * self.n_v
        for i in range(self.n_v):
            if not visited[i]:
                dfs_helper_it(i, visited)
                
        return result
    
G = Graph(5)
G.add_edge(0,1)
G.add_edge(0,3)
G.add_edge(1,2)
G.add_edge(3,2)
G.add_edge(3,4)

print("Bread FIrst Search => ", G.BFS(0))
print("Depth First Search => ", G.DFS())
print("Depth First Search Iterative", G.DFS_iter())
```
