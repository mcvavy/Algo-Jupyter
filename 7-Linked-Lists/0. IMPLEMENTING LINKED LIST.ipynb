```python
class LinkedList(object):
    class Node(object):
        def __init__(self, data):
            self.data = data
            self.next = None
            
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def __str__(self):
        current_node = self.head
        res = ""
        while current_node:
            res += str(current_node.data) + "-->"
            current_node = current_node.next
        return res
    
    def add_first(self, data: int) -> None:
        if self.head is None:
            self.head = Node(data)
            self.tail = self.head
        else:
            temp = self.head
            self.head = Node(data)
            self.head.next = temp
        self.size += 1
            
    def add_last(self, data: int) -> None:
        if self.head is None:
            self.head = Node(data)
            self.tail = self.head
        else:
            tail = self.tail
            tail.next = Node(data)
            self.tail = tail.next
        self.size += 1
    
    def remove_first(self) -> int:
        data = -1
        if not self.head:
            return -1
            
        if self.head == self.tail:
            data = self.head.data
            self.head = None
            self.tail = None
        else:
            data = self.head.data
            self.head = self.head.next
        
        self.size -= 1        
        return data
    
    def remove_last(self) -> int:
        if not self.head:
            return -1
        
        data = -1
        
        if self.head == self.tail:
            data = self.remove_first()
        else:
            current_node = self.head
            while current_node.next != self.tail:
                current_node = current_node.next
            data = self.tail.data
            self.tail = current_node
            current_node.next = None
            
        self.size -= 1
        return data
    
    def remove(self, data: int) -> int:
        if not self.head:
            return -1
        
        res = -1
        if self.head.data == data:
            res = self.remove_first()
        elif self.tail.data == data:
            res = self.remove_last()
        else:
            prev = None
            current = self.head
            
            while current and current.data != data:
                prev = current
                current = current.next
            
            if prev == self.tail:
                res = -1
            else:
                res = current.data
                prev.next = current.next
                self.size -= 1
        return res
    
    def contains(self, data):
        res = False
        if not self.head:
            res = False
        elif self.head.data == data or self.tail.data == data:
            res = True
        else:
            current = self.head
            while current and current.data != data:
                current = current.next
            
            if not current:
                res = False
            else:
                res = True
        return res
    
    def peek_first(self):
        if not self.head:
            return -1
        else:
            return self.head.data
    
    def peek_last(self):
        if not self.tail:
            return -1
        else:
            return self.head.tail
        
    def reverse(self):
        prev = None
        next_node = current = self.head
        
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev
            
    
linked_list = LinkedList()

linked_list.add_last(101)
linked_list.add_first(1)
linked_list.add_first(2)
linked_list.add_first(3)
linked_list.add_first(4)
linked_list.add_first(5)
linked_list.add_last(9)
linked_list.add_last(10)
linked_list.add_first(11)
linked_list.add_last(21)
linked_list.add_last(22)
    
# head = linked_list.head

# while head is not None:
#     print(str(head.data) + "-->")
#     head = head.next

print("Linked list size is ", str(linked_list.size))
print("Linked list data is ", str(linked_list))

print("===================================")
print("Remove first returns "+ str(linked_list.remove_first()))
print("Linked list size is ", str(linked_list.size))
print("Linked list data is ", str(linked_list))

print("===================================")
print("Remove last returns "+ str(linked_list.remove_last()))
print("Linked list size is ", str(linked_list.size))
print("Linked list data is ", str(linked_list))

print("===================================")
print("Remove 101 returns "+ str(linked_list.remove(101)))
print("Linked list size is ", str(linked_list.size))
print("Linked list data is ", str(linked_list))

print("===================================")
print("List contains 21? returns ", linked_list.contains(21))
print("List contains 101? returns ", linked_list.contains(101))


"""
Reversing a linked list
"""
print("=============Before reversing==============")
print("Linked list data is ", str(linked_list))

# ll = []
# current = linked_list.head

# while current:
#     ll.append(current)
#     current = current.next

# left = 0; right = len(ll) - 1
# while left < right:
#     ll[left].data, ll[right].data = ll[right].data, ll[left].data
#     left,right = left + 1, right - 1
    
linked_list.reverse()
print("=============After reversing==============")
print("Linked list data is ", str(linked_list))

    
```

> **Implementation of a Doubly linked list from scratch**

```python
class DoublyLinkedList(object):
    class Node(object):
        def __init__(self, data):
            self.data = data
            self.prev = prev
            self.next = next
    
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def __str__(self)-> str:
        current_node = self.head
        res = ""
        while current_node:
            res += str(current_node.data) + "-->"
            current_node = current_node.next
        return res
    
    def add_first(self, data: int) -> None:
        if not self.head:
            self.head = self.tail = Node(data)
            self.head.next = self.tail
            self.tail.prev = self.head
        else:
            new_node = Node(data)
            self.head.prev = new_node
            new_node.next = self.head
            self.head = self.head.prev
        self.size += 1
            
            
    def add_last(self, data: int) -> None:
        if not self.head:
            self.head = self.tail = Node(data)
            self.head.next = self.tail
            self.tail.prev = self.head
        else:
            new_node = Node(data)
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = self.tail.next
        self.size += 1
            
    
    def remove_first(self) -> int:
        if not self.head:
            return -1

        res = -1
        if self.head is self.tail:
            res = self.head.data
            self.head = self.tail = None
        else:
            res = self.head.data
            self.head = self.head.next
            self.head.prev = None
        self.size -= 1
        return res
            
    
    def remove_last(self) -> int:
        if not self.head:
            return -1

        res = -1
        if self.head is self.tail:
            res = self.head.data
            self.head = self.tail = None
        else:
            res = self.tail.data
            self.tail = self.tail.prev
            self.tail.next = None
        self.size -= 1
        return res
    
    def remove(self, data: int) -> int:
        pass
    
    def contains(self, data):
        pass
    
    def peek_first(self):
        pass
    
    def peek_last(self):
        pass
        
    def reverse(self)-> None:
        if not self.head:
            return
        
        head = self.head; tail = self.tail
        
        while head is not tail:
            head.data, tail.data = tail.data, head.data
            head = head.next; tail = tail.prev
    

double_linked_list = DoublyLinkedList()

double_linked_list.add_last(101)
double_linked_list.add_first(1)
double_linked_list.add_first(2)
double_linked_list.add_first(3)
double_linked_list.add_first(4)
double_linked_list.add_first(5)
double_linked_list.add_last(9)
double_linked_list.add_last(10)
double_linked_list.add_first(11)
double_linked_list.add_last(21)
double_linked_list.add_last(22)
    
# head = linked_list.head

# while head is not None:
#     print(str(head.data) + "-->")
#     head = head.next

print("Linked list size is ", str(double_linked_list.size))
print("Linked list data is ", str(double_linked_list))

print("===================================")
print("Remove first returns "+ str(double_linked_list.remove_first()))
print("Linked list size is ", str(double_linked_list.size))
print("Linked list data is ", str(double_linked_list))

print("===================================")
print("Remove last returns "+ str(double_linked_list.remove_last()))
print("Linked list size is ", str(double_linked_list.size))
print("Linked list data is ", str(double_linked_list))

print("=============Before reversing==============")
print("Linked list data is ", str(double_linked_list))
double_linked_list.reverse()
print("=============After reversing==============")
print("Linked list data is ", str(double_linked_list))
        
```
