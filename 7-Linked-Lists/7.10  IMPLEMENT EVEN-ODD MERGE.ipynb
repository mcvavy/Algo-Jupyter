> **Write a program that computes the even-odd merge.**  

_Hint: Use temporary additional storage._




```python
def even_odd_merge(L: ListNode) -> Optional[ListNode]:
    if not L:
        return L
    
    even_list = ListNode()
    even_pointer = even_list
    
    odd_list = ListNode()
    odd_pointer = odd_list
    
    pos = 0
    while L:
        if pos % 2 == 0:
            even_pointer.next = L
            even_pointer = even_pointer.next
        else:
            odd_pointer.next = L
            odd_pointer = odd_pointer.next
        L = L.next
        pos += 1
        
    even_pointer.next = odd_list.next
    return even_list.next
```
